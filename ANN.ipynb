```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
data.info()
```

```python
X = data.iloc[:, 3:13]
y = data.iloc[:, 13]
```

```python
geography = pd.get_dummies(X["Geography"], drop_first=True)
gender = pd.get_dummies(X['Gender'], drop_first=True)
```

```python
X = pd.concat([X, geography, gender], axis=1)
```

```python
X.head()
```

```python
X = X.drop(['Geography', 'Gender'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dropout
from keras.layers import Dense
```

```python
#initializing the NN
classsifier = Sequential()
```

```python
#adding first hidden layer
classsifier.add(Dense(units=6, kernel_initializer='he_uniform', activation='relu', input_dim=11))
```

```python
#addind the second hidden layer
classsifier.add(Dense(units=6, kernel_initializer='he_uniform', activation='relu'))
```

```python
#adding the final layer
classsifier.add(Dense(units=1, kernel_initializer='glorot_uniform', activation='sigmoid'))
```

```python
#summary
classsifier.summary()
```

```python
#compiling ANN
classsifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
#fitting
model_history = classsifier.fit(X_train, y_train, validation_split=0.33, batch_size=10, nb_epoch=100)
```
